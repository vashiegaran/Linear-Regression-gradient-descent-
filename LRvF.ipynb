```python
import numpy as np
import pandas as pd
```

```python
dataset_path = './dataset/dataset_csv.csv'
num_epoch=2000
batch_size=1000
Learning_rate= 0.00001
```

```python
df = pd.read_csv(dataset_path) #read the dataset
df.head()
```

```python
X= df.iloc[: ,:-1]# To sort all the independept Data into X variable
Y=df.iloc[: ,-1]# to sort all the dependant data into Y variable
```

```python
b = np.ones([X.shape[0],X.shape[1]+1]) #Adding an extra column with values of 1
```

```python
b[: , 1:] = X # merging the real datas with the one with extra column
```

```python
X=b
```

```python
theta = np.random.rand(X.shape[1]) #randomly assign a floating values to each independant data's column
```

```python
X_train= X[:-1000 , :]# Assing row 0 until last 1000 to X_train variable
Y_train = Y[:-1000]
X_test = X[-1000:, :]
Y_test = Y[-1000:]
```

```python
for epoch in range (num_epoch): #Iterate the process num_epoch times
    Y_diff = X_train.dot(theta) - Y_train
    loss_func = 1/(2*X.shape[0]) * (Y_diff.dot(Y_diff))# Loss function
    
    for first_index in range (0,X.shape[0],batch_size):# Process the first batch_size data 
        last_index = first_index + batch_size# to assign last_index value
        last_index = None if last_index > X_train.shape[0] else last_index# if the values become more than the number of row which is values it will take the last number of rows
        
        
        Y_pred = X_train[first_index:last_index].dot(theta) # to predict the y value  based on the theta for according to the batch_size
        y_diff = Y_pred - Y_train[first_index:last_index] # to find the difference between the predicted y value and the real y value
        
        m=1/Y_train.shape[0]
        
        for i in range(theta.shape[0]):# to iterate through every column to find the new theta
            part_derv= m*y_diff.dot(X_train[first_index:last_index,i]) # perform partial derivative to find theta where it minimise the loss function
            theta[i] = theta[i] - Learning_rate*part_derv # to find a new theta where the loss function is minimum
            
            
    print("Epoch :", epoch)
    print("Loss :", loss_func)

    
```
